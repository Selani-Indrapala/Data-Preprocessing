```python
import numpy as np
import matplotlib.pyplot as plt

def generate_signal(signal_length, num_nonzero):
    
    signal = np.zeros(signal_length)
    nonzero_indices = np.random.choice(signal_length, num_nonzero, replace=False)
    nonzero_values = 50*np.random.randn(num_nonzero)  
    signal[nonzero_indices] = nonzero_values
    return signal

signal_length = 100  # Total length of the signal
num_nonzero = 10    # Number of non-zero elements in the signal
your_index_no= 200232 # Enter without english letter and without leading zeros
signal = generate_signal(signal_length, num_nonzero)
signal[10] = (your_index_no % 10)*10 + 10
if  your_index_no % 10 == 0:
  signal[10] = np.random.randn(1) + 30


signal=signal.reshape(signal_length,1)
plt.figure(figsize=(15,5))
plt.subplot(1, 1, 1)
plt.title("Data")
plt.stem(signal)
```

```python
#Implementing MaxAbsScalar
from sklearn.preprocessing import MaxAbsScaler

scaled_data_MaxAbsScalar = MaxAbsScaler().fit_transform(signal)
plt.figure(figsize=(15,5))
plt.subplot(1, 1, 1)
plt.title("MaxAbsScalar")
plt.stem(scaled_data_MaxAbsScalar)
```

```python
def min_max_scale(data):
    min_val = np.min(data)
    max_val = np.max(data)
    print("min of data", min_val, "max of data", max_val)
    scaled_data = (data - min_val) / (max_val - min_val)
    return scaled_data


scaled_data_min_max = min_max_scale(signal)

plt.figure(figsize=(15,5))
plt.subplot(1, 1, 1)
plt.title("MinMax Scale")
plt.stem(scaled_data_min_max)
```

```python
def std_normalization_scale(data):
    mean = np.mean(data)
    std_deviation = np.std(data)
    scaled_data = (data-mean)/std_deviation
    return scaled_data


scaled_data_std_normalization = std_normalization_scale(signal)

plt.figure(figsize=(15,5))
plt.subplot(1, 1, 1)
plt.title("Standard Normalization Scale")
plt.stem(scaled_data_std_normalization)
```

```python
#Non Zero Elements

def Non_Zero(data):
    count = 0
    for d in data:
        if (d!=0):
            count +=1
    return count

print("Non-zero data in Original signal: ",Non_Zero(signal))
print("Non-zero data in MaxAbsScalara signal: ",Non_Zero(scaled_data_MaxAbsScalar))
print("Non-zero data in MinMaxScalar signal: ",Non_Zero(scaled_data_min_max))
print("Non-zero data in Standard Normalised signal: ",Non_Zero(scaled_data_std_normalization))
```
